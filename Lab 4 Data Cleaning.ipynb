## Dataset Preparation 1 (Attendance)

My first move is to copy our attendance dataset to a new variable, as this will allow me to edit it without making permanent changes. 

```python
qc_dfa = dfa
```

Now that we have copied our dataset, we want to know all the unique values for the first feature.

```python
qc_dfa['Table 43. Average daily attendance (ADA) as a percentage of total enrollment, school day length, and school year length in public schools, by school level and state: 2007-08'].unique()
```

From the output of the unique function, we can find out that our first row consists of each of the 50 states in addition to the District of Columbia and The United States as a whole. We want to save these values while getting rid of the null values including the dataset information at the bottom and top of the dataset.

```python
qc_dfa.dropna(inplace = True)
```

This code simply deletes all rows that consist of null values. This is why we copied our dataset. Ultimately I chose to drop all null values because the null values dropped only included extra information and did not delete any data on the 50 states, the United States, or the District of Columbia, which are the rows I want to examine. Given this, deleting all other data cleans up the dataset while not hurting the data we need and love. 

```python
qc_dfa['Table 43. Average daily attendance (ADA) as a percentage of total enrollment, school day length, and school year length in public schools, by school level and state: 2007-08'].unique()
```

Here we check to make sure we deleted all of the rows we did not want, while keeping the rows we do want. Since this is true, we can continue to clean.

Our next step is to create new features. 

```python
new_features = ['State','ADA (Combined)','delete1','Average Hours in School Day (Combined)','delete2','Average Days in School Year (Combined)', 'delete3','Average Hours in School Year (Combined)','delete4','ADA (Elementary)','delete5','Average Hours in School Day (Elementary)','delete6','ADA (Secondary)','delete7','Average Hours in School Day (Secondary)','delete8']
```

First we create a new list of all our new feature names, since some of our columns do not have headers, we will mark them with delete so we know to delete them later. I made this decision because ultimately if we do not know what the data is, it serves us no purpose and it would be counter-productive to make assumptions. By simply marking them for deletion and then later deleting them, I am left with only the data I know. 

```python
df_b = pd.read_excel("attendance.xls", #name of csv file
                 header=3,         #which indexed row in the file is the last header row this means rows 0..3 are headers
                 names=new_features, #use the new_labels list for the column names. Notice there are no blanks
                 index_col='State')  # use the last_name column as our index
```

We read in the attendance file again but this time with our new headers and our index being the State name. By setting the index to be the State name, we are allowing the user to A) more easily understand the data is for each state, and B) allow them to search by state name more easily for the data attached to that state.

```python
df_b.dropna(inplace = True)
```

We again want to drop all unnecessary data and all null values, but this time with our index being the state name and our headers being our new headers.

```python
df_b
```

Here we have a fairly clean dataset but we can still do more!

```python
df_b.drop(['delete1','delete2','delete3','delete4','delete5','delete6','delete7','delete8'], axis=1, inplace = True)
```

Since we do not know what those columns are and they have no explanation, I made the decision to delete them since I cannot title them without knowing what they do. Ultimately having data without any knowledge of the data means or  is of, is unhelpful. 

```python
df_b
```

Next we want to clean up the State names, since they are currently in an index, we first have to remove them from the index. Creating a new index does this.

```python
df_b = df_b.reset_index()
```

```python
df_b
```

Then lets get rid of any periods from the State. This will allow people to more easily find the States they need and search for them, without having to guess how many periods are after it.

```python
df_b["State"] = df_b["State"].str.strip('.')
```

We then want to strip any spaces at the ends of the State names, this way they will be even more searchable.

```python
df_b["State"] = df_b["State"].str.strip()
```

```python
df_b
```

Lastly, lets set the index to be equal to the State again. This way people can view the data more easily.

```python
df_b.set_index('State', drop=True, append=False, inplace=True, verify_integrity=False)
```

```python
df_b
```

```python
df_b.describe()
```

Looking at our data set we see it is perfectly clean! We left a few non-numerical values inside the dataset because ultimately these serve to tell a story, one where just by looking at the dataframe you can make assumptions or predictions about certain privileges or restrictions in certain states or areas. Ultimately this is another part of the data and one I believe is important to preserve.
